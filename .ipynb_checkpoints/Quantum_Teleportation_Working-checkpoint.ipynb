<h1> Circuit for Quantum Teleporation between Sender and Reciever </h1>

<h3> Here we are teleporting a quantum state <b>|$\Psi$></b> from our senders ip address to receiever ip address </h3>


1.) First, Sender and Reciever form an entangled <b>Bell pair</b> between their two qubits.
It is accomplished through the use of a Hadamard and CX gate. 

2.) Following that, Sender performs a series of operations on the qubit with state |$\Psi$> and their half of the Bell pair. 

3.) Finally, Sender measures both qubits (the one that initially contained |$\Psi$> and the one that contained her part of the Bell pair) and communicates the results to Recievers.  

4.) Reciever applies a couple of gates to his part of the Bell pair depending on what Sender says. 

5.) <center><b>Drumroll!!</b></center>  <h4><center> lo and behold! The state has reached at receivers qubit. :D</center><h4><center><b>!!Drumroll</b></center>

```python
qcirc = QuantumCircuit(3, 3)

qcirc.x(0)
qcirc.barrier()

qcirc.h(1)
qcirc.cx(1, 2)
qcirc.barrier()

qcirc.cx(0, 1)
qcirc.h(0)

qcirc.measure([0, 1], [0, 1])
qcirc.barrier()

qcirc.cx(1, 2)
qcirc.cz(0, 2)

qcirc.measure([2], [2])
```

<h1> Quantum teleportaion Simulation </h1>

<h4> We will be running the circuit on Qasm Simulator which is a simulator not an actual Quantum Computer as simulator are currently faster and more precise that we require for this demonstration </h4>

```python
sim = Aer.get_backend('qasm_simulator')
outcome = execute(qcirc, backend=sim, shots=1024).result()
from qiskit.visualization import plot_histogram
plot_histogram(outcome.get_counts(qcirc))
```

Here we are seeing a plot where the order is <b>qubit2</b> <b>qubit1</b> <b>qubit0</b>. As we can see from outcome <b>qubit2</b> will be always be equal to unity we teleported the state <b>|1></b> to <b>qubit2</b>

<h3> To make this simulation more clearer here is the circuit explaining above process </h3>

```python
def build_psi(angle):
    qcirc = QuantumCircuit(1, name='build_psi')
    qcirc.h(0)
    qcirc.rz(angle, 0)
    return qcirc

def Bell_pair():
    qcircBell = QuantumCircuit(2, name='Bell pair constructed')
    qcircBell.h(0)
    qcircBell.cx(0,1)
    return qcircBell

def SendersTeleportation():
    senderqcirc = QuantumCircuit(2, name='Sender Transportation')
    senderqcirc.cx(0,1)
    senderqcirc.h(0)
    return senderqcirc

def senderMeasurement():
    senderqcirc = QuantumCircuit(2, 2, name='Senders Measurement process')
    senderqcirc.measure([0,1], [0,1])
    return senderqcirc

def RecieversTeleportation():
    recieverqcirc = QuantumCircuit(3, name='Reciever Transportation')
    recieverqcirc.cx(1,2)
    recieverqcirc.cz(0,2)
    return recieverqcirc

def build_circuit(angle):
    qcirc = QuantumCircuit(3, 3)
    qcirc.append(build_psi(angle).to_instruction(), [0])
    qcirc.barrier()
   
    qcirc.append(Bell_pair().to_instruction(), [1,2])
    qcirc.barrier()
   
    qcirc.append(SendersTeleportation().to_instruction(), [0,1])
    qcirc.barrier()
    
    qcirc.append(senderMeasurement().to_instruction(), [0,1], [0,1])
    qcirc.barrier()
    
    qcirc.append(RecieversTeleportation().to_instruction(), [0,1,2])
    qcirc.barrier()
    
    qcirc.h(2) 
    qcirc.measure([0,1,2], [0,1,2])
    return qcirc

qcirc = build_circuit(0.01)
qcirc.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
outcome = execute(qcirc, backend=simulator, shots=1024).result()
counts = outcome.get_counts(qcirc)
print(counts)
num_c2_zero = sum(counts[c2c1c0] for c2c1c0 in counts if c2c1c0[0] == '0')

```

<h2> Graph for different $\theta$ </h2>

```python
import numpy as np
angles = np.arange(0, 4*np.pi, np.pi/16)

sim = Aer.get_backend('qasm_simulator')

percent_ones = []

for angle in angles:
    qcirc = build_circuit(angle)
    outcome = execute(qcirc, backend=simulator, shots=1024).result()
    counts = outcome.get_counts(qcirc)
    num_c2_ones = sum(counts[c2c1c0] for c2c1c0 in counts if c2c1c0[0] == '1')
    percent_ones.append(num_c2_ones*100./1024)
    
import matplotlib.pyplot as plotter
plotter.plot(angles, percent_ones)
plotter.xlabel('Initial angle, theta (radians)')
plotter.ylabel('Percentage of 1 counts')
plotter.show()
```

<h1> We will run above teleportation circuit on real quantum computer </h1>

```python
angles = np.arange(0, 4*np.pi, np.pi/16)

qcircs = []
for angle in angles:
    qcirc = build_circuit(angle)
    qcircs.append(qcirc)
```
